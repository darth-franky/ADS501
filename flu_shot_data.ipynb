# Flu Shot Data

## Load the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('test_set_features.csv')

print(df.head())

print("\nDataset Info:")
print(df.info())
```

## Data Exploration

### Visualization 1: H1N1 Concern Levels Distribution

```python

# Distribution of H1N1 Concern Levels
plt.figure(figsize=(8, 5))
df['h1n1_concern'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of H1N1 Concern Levels')
plt.xlabel('Concern Level (0: None, 1: Low, 2: Moderate, 3: High)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Visualization 2: Distribution of Age Groups

```python
plt.figure(figsize=(8, 5))
df['age_group'].value_counts().sort_index().plot(
    kind='bar', 
    color='lightgreen', 
    edgecolor='black'
)
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Data Quality Report

### Missing Values

```python
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_report = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percentage
}).sort_values(by='Missing Values', ascending=False)

print("\nMissing Values Report:")
print(missing_report)
```

### Data Summary Report

```python
data_summary = pd.DataFrame({
    'Column': df.columns,
    'Data Type': df.dtypes.values,
    'Unique Values': [df[col].nunique() for col in df.columns],
    'Non-Null Count': df.notnull().sum().values,
    'Null Count': df.isnull().sum().values
})

print("\nData Summary Report:")
print(data_summary)
```

# Insights

## Visualizations Insights:
- **H1N1 Concern Levels:** Most respondents showed low to moderate concern, which may affect vaccination rates.  
- **Age Group Distribution:** Respondents are mostly between 45-64 years, suggesting a critical target group for vaccination programs.  

## Data Quality Report Insights:
- **Missing Data:**  
  - `health_insurance`: 45.8% missing, may need imputation or removal.  
  - `employment_industry` and `employment_occupation`: Over 50% missing, could affect employment-based analysis.  
- **Data Types:** All columns are correctly typed as numeric or categorical.
